```python
import random
import matplotlib.pyplot as plt
from algorithm import Bees  
from generator import Generator 
```

```python
def print_pink(value):
    return f"\033[95m{value}\033[0m"

def run_algorithm_tests_and_plot_results():
    test_sizes = list(range(5, 105, 5))  
    algorithm_results = []
    random_results = []

    for size in test_sizes:
        size_text = print_pink(size)
        size_times_2_text = print_pink(size * 2)
        size_div_2_text = print_pink(size // 2)
        
        print(f'Running test with data size: {size_text}, {size_times_2_text}, {size_div_2_text}')
        
        random_data = generate_random_test_data(size)
        
        bees = Bees(20, random_data["airplanes"], random_data["passenger_groups"], random_data["airports"])
        final_solution, final_population, fitness_new = bees.bees_algorithm(random_data)
        
        random_solution = bees.generate_random_solution()
        random_solution_evaluation = bees.evaluate_solution(random_solution)
        
        algorithm_results.append(fitness_new)
        random_results.append(random_solution_evaluation)

    plt.plot(test_sizes, algorithm_results, label='Wynik algorytmu')
    plt.plot(test_sizes, random_results, label='Wynik losowego dopasowania')
    plt.xlabel('Rozmiar danych')
    plt.ylabel('Otrzymany Zysk')
    plt.title('Porównanie zysku otrzymanego przez algorytm i losowe dopasowanie')
    plt.legend()
    plt.show()

def generate_random_test_data(size):
    return Generator.generate_random_test_data(size, size * 2, size // 2)

run_algorithm_tests_and_plot_results()
```
